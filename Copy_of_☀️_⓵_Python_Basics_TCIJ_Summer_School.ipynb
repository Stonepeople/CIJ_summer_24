\# **For loops, range and append combined in a simple scraper**

For loops are very useful, and combined with range they really come into their own because we can do things many times very quickly. In this example it allows us to build up a list of webpages. Run the example below:

````python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd

for page in range(1,10):
    URL = "https://www.theguardian.com/news/series/todayinfocus?page=" + str(page)

    print(URL)
````

We can take that a step further and use a simple scraper to scrape each page to get all the podcast headlines

We'll look at this code again in lesson 2 on scrapers !

````python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd

# data = [] — an empty list/array, ready to collect results
data = []

# the outer for loop visits each of the 5 Guardian webpages, one at a time

for webpage_no in range(1,6):
    URL = "https://www.theguardian.com/news/series/todayinfocus?page=" + str(webpage_no)

    website_request = requests.get(URL)
    website_content = website_request.text
    website_read = bs(website_content)

    # The inner loop finds every matching headline on that one page
    # .select() returns a list of matches, so this inner loop is going through each one
    # data.append(headline.text()) — for each headline found, grab its text and add it onto the end of data.

    for headline in website_read.select("h3.card-headline span"):
        data.append(headline.text)

print(data)
````
